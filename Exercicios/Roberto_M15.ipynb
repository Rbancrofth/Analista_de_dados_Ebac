## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
with sns.axes_style('whitegrid'):

  grafico = sns.boxplot(x=data["price"])
  grafico.set(title='Distribuição de Preço', xlabel='USD');
```

**Insight do gráfico 1**: ...

Analisando o gráfico, percebemos uma concentração significativa dos preços de diamantes na faixa entre US$1250 e US$5000. A partir de US$12000, a frequência dos valores diminui consideravelmente, indicando que diamantes nessa faixa de preço são menos encontrados no mercado.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = data[['weight', 'model_year']]
cars.head()
```

```
with sns.axes_style('whitegrid'):
  sns.lmplot(x="model_year", y="weight", data=cars, scatter=False)

plt.title("Tendência do peso dos carros ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Peso do carro (lb)")

# Mostrar o gráfico
plt.show()
```

**Insight do gráfico 3**: ...

A partir da análise do gráfico, é possível inferir que o peso médio dos automóveis apresentou uma decréscimo gradual e ininterrupto durante o período compreendido entre 1970 e 1982.
